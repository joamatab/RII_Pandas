# Example

```
import numpy as np
import matplotlib.pyplot as plt
import riip
ri = riip.RiiDataFrame()
plt.style.use('seaborn-poster')
props = plt.rcParams['axes.prop_cycle']
```

## Ag

```
ri.search_pages('Ag')
```

```
Ag_idx_list = [0, 1694, 1697, 1698, 1699, 1711]
ri.show(Ag_idx_list)
```

### Dielectric function

```
alpha = 0.6
wl_min = 0.25
wl_max = 1.0
wls = np.linspace(wl_min, wl_max, 200)
for idx, prop in zip(Ag_idx_list, props):
    Ag = riip.Material(ri.catalog.loc[idx], ri.raw_data.loc[idx])
    eps = Ag.eps(wls)
    plt.plot(wls, eps.real, "-", color=prop['color'], alpha=alpha, lw=4,
             label="{}".format(ri.catalog.loc[idx, 'page']))
    plt.plot(wls, eps.imag, "--", color=prop['color'], alpha=alpha, lw=4)
plt.xlim(wl_min, wl_max)
plt.ylim(-50, 10)
plt.legend(loc='lower left')
plt.show()
```

## Water

```
ri.search_pages('H2O')
```

### Wavelength dependence of refractive index

```
water_idx_list = [282, 283, 284, 285]
alpha = 0.6
wl_min = 0.38
wl_max = 0.75
wls = np.linspace(wl_min, wl_max, 200)
for idx, prop in zip(water_idx_list, props):
    water = riip.Material(ri.catalog.loc[idx], ri.raw_data.loc[idx])
    plt.plot(wls, water.func_n(wls), "-", color=prop['color'], alpha=alpha, lw=4,
             label="{}".format(ri.catalog.loc[idx, 'page']))
plt.xlim(wl_min, wl_max)
plt.ylim(1.325, 1.35)
plt.legend(loc='lower left')
plt.show()
```

### Averaged refractive index

```
wls = np.linspace(0.38, 0.75, 200)
water = riip.Material(ri.catalog.loc[283], ri.raw_data.loc[283])
np.average(water.func_n(wls))
```

### Extinction coefficient

```
idx = 281
alpha = 0.6
wl_min = 0.5
wl_max = 1.6
wls = np.linspace(wl_min, wl_max, 200)
water = riip.Material(ri.catalog.loc[idx], ri.raw_data.loc[idx])
plt.plot(wls, water.func_k(wls), "-", alpha=alpha, lw=4,
         label="{}".format(ri.catalog.loc[idx, 'page']))
plt.xlim(wl_min, wl_max)
plt.legend(loc='upper left')
plt.show()
```
